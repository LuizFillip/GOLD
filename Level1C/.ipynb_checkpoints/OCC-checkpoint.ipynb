```python
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
import numpy as np 
import cartopy.crs as ccrs
import cartopy.feature as cf

from split_filename import *
from GOLD import *
import os
```

```python
infile = 'C:\\Users\\Luiz Felipe\\Downloads\\OCC2019111120191112\\tmp\\archive_L1C\\2019\\315\\' 

filename = 'GOLD_L1C_CHA_OCC_2019_315_06_23_v02_r01_c01.nc'

ds = xr.open_dataset(infile + filename)

ds.coords['n_wavelength'] = ds['WAVELENGTH'].values[0]

def bytes_string(a):
    func = np.vectorize(lambda a: pd.to_datetime(a.decode("utf-8")))
    return xr.apply_ufunc(func, a)

ds.coords['nTimes'] = bytes_string(ds['TIME_UTC'])



ds
```

```python
from glob import glob

paths = sorted(glob(infile +  '*.nc'))

datasets_in_list = [xr.open_dataset(filename) for filename in paths]

datasets = xr.concat(datasets_in_list, dim = 'nTimes')
               
```

```python
datasets.coords['nTimes'] = bytes_string(datasets['TIME_UTC'])

datasets.coords['n_wavelength'] = datasets['WAVELENGTH'].values[0]

array_wave = datasets['WAVELENGTH'].values[0]

cond = (array_wave > 133) & (array_wave < 137)
        
result_1356 = np.where(cond, array_wave, np.nan)

result_1356 = result_1356[~np.isnan(result_1356)]

result_1356

ds = datasets.sel(n_wavelength = result_1356)#.mean(dim = 'n_wavelength')

ds
```
